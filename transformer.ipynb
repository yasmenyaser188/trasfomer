```python
import torch
import torch.nn.functional as F

# Define the input sentence
sentence = ["What", "are", "the", "symptoms", "of", "diabetes?"]
vocab_size = 100  # Arbitrary vocabulary size
embedding_dim = 8  # Embedding size

# Simulate word embeddings (random for demonstration)
torch.manual_seed(42)
embeddings = torch.rand(len(sentence), embedding_dim)

# Define trainable query, key, and value matrices
d_k = embedding_dim  # Dimension of keys/queries
W_q = torch.rand((embedding_dim, d_k))
W_k = torch.rand((embedding_dim, d_k))
W_v = torch.rand((embedding_dim, d_k))

# Compute Q, K, V matrices
Q = embeddings @ W_q  # Query matrix
K = embeddings @ W_k  # Key matrix
V = embeddings @ W_v  # Value matrix

# Compute Attention Scores
attn_scores = (Q @ K.T) / (d_k ** 0.5)  # Scaled dot-product
attn_weights = F.softmax(attn_scores, dim=-1)  # Apply softmax

# Print Attention Scores
print("Self-Attention Scores:\n", attn_weights)
```
